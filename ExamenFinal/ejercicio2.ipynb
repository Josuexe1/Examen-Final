```python
from scipy import optimize
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return (x**5-6*x**4+2*x**3+20*x**2-27*x+10)
    fprime = lambda x: 5 * x**4 - 24 * x**3 + 6 * x**2 + 40 * x - 27
    roots = [-2.0, 5.0, 1.0000040198387017, 0.9999931250540883]
    unique_roots = np.unique(roots)
    sorted_roots = np.full(5, np.nan)
    sorted_roots[:len(unique_roots)] = np.sort(unique_roots)
    sorted_roots
```

```python
x = np.linspace(-10, 10, 400)
y = f(x)
plt.plot(x, y, label='y = x^5-6x^4+2x^3+20x^2-27x+10')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Gráfica de la función")
plt.legend()
plt.grid(True)
plt.show()
```
